````python
import re

def file_reader(filename):
    words = [line.strip() for line in open(filename)]
    bigrams = []
    for ln in words :
           text = re.sub('\W', ' ', ln.lower())
           words = text.split()
           bigrams.append(list(zip(words, words[1:])))
    return (bigrams)

bigrams = file_reader(r"C:\Users\COLLEDANI Romain\Documents\COURS\CENTRALE\Big DATA\Assignement 1\assignment_1_prices\data\ulysses_extract.txt")
print(bigrams)
````

## Part 1 - unique bigrams
Describe and define a Map Reduce procedure that gives as output the number of unique bigrams that appear in the document.

You can use any data structure you want, here a possible implementation of a function that reads the document and give as output the list of bigram in the text is provided. 

Notice that this function reads the bigrams of an input file and stores them in a list.  Feel free to re-implement your read function

````python
from collections import defaultdict

#the function map_bigrams take as a parametre a file and return a list. Each element of this list is a list of all the 
#bigrams in a given line (with the number 1 in front of each bigram).

def map_bigrams(filename):
    bigrams = file_reader(filename)
    res=[]
    for line in bigrams:
        lin=[]
        for bigram in line:
            lin.append([bigram,1])
            if (len(lin)!=0): 
                res.append(lin)
    return res
   
mapp = map_bigrams(r"C:\Users\COLLEDANI Romain\Documents\COURS\CENTRALE\Big DATA\Assignement 1\assignment_1_prices\data\ulysses_extract.txt")     
print(mapp)

#the function suffle_bigrams take as a parametre the output of the previous function and return a dictionnary.
#the keys of this dictionnary are all the bigrams encountered in the initial file, and the value for a given key is this 
#bigram.

   
def shuffle_bigrams(map_bigram):
    d=defaultdict(list)
    for big in map_bigram:
        for i in range (len(big)):
            d[big[i][0]].append(big[i])
            
        
    return d

shuff = shuffle_bigrams(mapp)
#print(shuff)
            

        
    


#the function reduce_bigrams take as a parametre the output of the previous function and return a list.
#each element of this list is a tuple. The first element of this tuple is a bigram and the second is the number of time 
#we meet this bigram in the initial text.

def reduce_bigrams(shuffle):
    res=[]
    for big in shuffle.keys():
        total=0
        for i in range (len(shuffle[big])):
            total+=shuffle[big][i][1]
        res.append((big,total))
    return res
    
    
   
reduce = reduce_bigrams(shuff)
#print(reduce)



     

````

````python
# After the mapping and before the shuffle, we can call this combine_bigrams fonction. This function directly return all the
#bigrams per line and the number of time we encoutered this bigram on this line. 

def  combine_bigrams(map_bigram):
    combine=[]
    for line in map_bigram:
        combine_line=[]
        visited_bigrams=[]
        for big in line:
            if big not in visited_bigrams:
                visited_bigrams.append(big)
                combine_line.append([big[0],line.count(big)])
        combine.append(combine_line)
    return (combine)


mapp = map_bigrams(r"C:\Users\COLLEDANI Romain\Documents\COURS\CENTRALE\Big DATA\Assignement 1\assignment_1_prices\data\ulysses_extract.txt")
comb = combine_bigrams(mapp)
shuff = shuffle_bigrams(comb)
reduce = reduce_bigrams(shuff)
print(reduce)
````

````python
#This function allows us to get a workable bill from a bill with the shape of the ticket from the file supermarket.txt

def list_supermarket_reader(filename):
    words = [line.strip() for line in open(filename)]
    lin=[]
    res=[]
    for ln in words:
        text=ln.split()
        lin.append(text)
    for lines in lin:
        tex=lines[0].split(',')
        res.append(tex)
    return res

#print(list_supermarket_reader(r"C:\Users\COLLEDANI Romain\Documents\COURS\CENTRALE\Big DATA\Assignement 1\assignment_1_prices\data\supermarket.txt"))

#This function take as a parametre a supermarket ticket (as a file) a return a list of tuple, each tuple et a year-month date
#and a article purchased at this periode.

def  map_super(filename):
    ticket = list_supermarket_reader(filename)
    res=[]
    for line in ticket:
        map_line=[]
        date_split = line[0].split("-")
        date=str(date_split[0])+"-"+str(date_split[1])
        for i in range (len(line)-1):
            map_line.append((date,line[i+1]))
        res.append(map_line)
    return res


mapp = map_super(r"C:\Users\COLLEDANI Romain\Documents\COURS\CENTRALE\Big DATA\Assignement 1\assignment_1_prices\data\supermarket.txt")
#print(mapp)
            

            

#the function suffle_super take as a parametre the output of the previous function and return a dictionnary.
#the keys of this dictionnary are the date "year-month" encountered in the initial file, and the value for a given key is the
#list of all the articles purchased during this period

def shuffle_super(mapp_super):
    d=defaultdict(list)
    for purchases in mapp_super:
        for i in range (len(purchases)):
            d[purchases[i][0]].append(purchases[i][1])
               
    return d

shuff = shuffle_super(mapp)
print(shuff)
            

        

#the function reduce_super take as a parametre the output of the previous function and return a list.
#each element of this list is an other list. The first element of this list is a given period (month-year) and the next 
#10 elements are the 10 most sold articles during this period with the number of sales.

def reduce_super(shuffle):
    res=[]
    for date in shuffle.keys():
        articles=[]
        articles_dict = defaultdict(int)
        for i in range (len(shuffle[date])-1):
            articles_dict[shuffle[date][i+1]]+=1
        
        for e in articles_dict.keys():
            articles.append((e,articles_dict[e]))
        articles = sorted(articles, reverse=True, key=lambda x: x[1])
        articles.insert(0,date)
        res.append(articles[:11])
    
    return res
    
    
   
reduce = reduce_super(shuff)
#print(reduce)
````

````python
#print(list_supermarket_reader(r"C:\Users\COLLEDANI Romain\Documents\COURS\CENTRALE\Big DATA\Assignement 1\assignment_1_prices\data\supermarket_prices.txt"))

# This function reuse the previous map_super but transform each element of a given line (except the date) in a tuple 
# (article, price)

def  map_income(filename):
    ticket = map_super(filename)
    new_ticket=[]
    for line in ticket:
        new_line=[]
        for article in line:
            article_price = article[1].split("-")
            article_price[1]=float(article_price[1])
            new_line.append((article[0],article_price[0],article_price[1]))
        new_ticket.append(new_line)
            
    return new_ticket
    
mapp_income = map_income(r"C:\Users\COLLEDANI Romain\Documents\COURS\CENTRALE\Big DATA\Assignement 1\assignment_1_prices\data\supermarket_prices.txt")
print(mapp)

#This function is exactly the same as the function shuffle_super, except that, for a given date, I add not only the article
#but also its price.

def shuffle_income(mapp_income):
    d=defaultdict(list)
    for purchases in mapp_income:
        for i in range (len(purchases)):
            d[purchases[i][0]].append((purchases[i][1],purchases[i][2]))
               
    return d

shuff_income= shuffle_income(mapp_income)
#print(shuff_income)  




#this function is exactly the same as reduce_super except that, for a given date, for a given article, I do not add "1" 
#(in order to count the number of article), but I add his price, in order to count the total amount received from the
#sale of this article.

def reduce_income(shuffle):
    res=[]
    for date in shuffle.keys():
        articles=[]
        articles_dict = defaultdict(float)
        for i in range (len(shuffle[date])-1):
            articles_dict[shuffle[date][i+1][0]]+= shuffle[date][i+1][1]
        
        for e in articles_dict.keys():
            articles.append((e,round(articles_dict[e],2)))
        articles = sorted(articles, reverse=True, key=lambda x: x[1])
        articles.insert(0,date)
        res.append(articles)
    
    return res
   
reduce=reduce_income(shuff_income)
#print(reduce)
                                      
````

````python
#This mappin function return a list of list. Each element of this last list is a tuple (ticket_id, (tuple of products))

def map_super_percentage(filename):
    ticket = list_supermarket_reader(filename)
    res=[]
    id=1
    for line in ticket:
        map_line=[]
        if (len(line)==2): #If there is only one article
            map_line.append((id,(line[1],'')))
        else:
            for i in range (1,len(line)-1):
                for j in range (i+1,len(line)):
                    map_line.append((id,(line[i],line[j])))
        res.append(map_line)
        id+=1
    return res

mapp_percent = map_super_percentage(r"C:\Users\COLLEDANI Romain\Documents\COURS\CENTRALE\Big DATA\Assignement 1\assignment_1_prices\data\supermarket.txt")
#print(mapp_percent)



#This shuffle function take as a parameter the mapping previously done, and return two dictionnaries. The first one keys are
#all the pairs of article purchased together (knowing that the pair (article_1,article_2) is the same as (aticle_2,article_1))
#and its values are the list of the distinct ticket_id where this pair is present.
#The second disctionnary keys are all the articles purchased and its values are lists of distinct ticket_id where this 
#artcile is present.

def shuffle_super_percentage(mapping):
    d_tuple=defaultdict(list)
    d_article=defaultdict(list)
    for purchase in mapping:
        for tupl in purchase:
            if ((tupl[1][1],tupl[1][0])) not in d_tuple.keys():
                d_tuple[tupl[1]].append(tupl[0])
            d_article[tupl[1][0]].append(tupl[0])
            d_article[tupl[1][1]].append(tupl[0])
    for article in d_article.keys():
        d_article[article]=list(set(d_article[article]))
    del d_article[""]
    return (d_tuple,d_article)

shuff_percent= shuffle_super_percentage(mapp_percent)
#print(shuff_percent[0])  
    
            
#This function takes as parameters the shuffle previously done, and the number of ticket (by default=100)
#As we remove the redundants tickets in the prievious shuffle function (for both dictionnary, we have only the distinct
# ticket_id), we just have to count the number of ticket where we can see each tuple and each article. 


def reduce_super_percentage(shuffle, nb_ticket=100):
    res = []
    for e in shuffle[0].keys():
        percent_tuple= (len(shuffle[0][e])/nb_ticket)*100
        percent_article_1 = (len(shuffle[1][e[0]])/nb_ticket)*100
        percent_article_2 = (len(shuffle[1][e[1]])/nb_ticket)*100
        res.append((e,round(percent_tuple,1),round(percent_article_1,1),round(percent_article_2,2)))
    res = sorted(res, reverse=True, key=lambda x: x[1])
    return res

reduce = reduce_super_percentage(shuff_percent)
print(reduce)
````
